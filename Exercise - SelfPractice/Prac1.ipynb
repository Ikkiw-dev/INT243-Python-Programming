```python
#4
scores = [45, 78, 90, 66, 82]

maxsc = scores[0]

for i in scores:
    if i > maxsc:
        maxsc = i

minsc = scores[0]
for j in scores:
    if j < minsc:
        minsc = j

acc = 0
for k in scores:
    acc += k

avg = acc / len(scores)

print(f"max {maxsc}, min {minsc}, total {acc}, average {avg}")
```

```python
#6 (More Professional)
def get_number(count = 5):
    number = []
    while len(number) < count:
        try:
            prefer_num = float(input(f"Enter number {len(number) + 1} : "))
            number.append(prefer_num)
            print(f"Your actual number {number}")
        except ValueError:
            print(f"Invalid Values, Please try again later")
    return number
def main():
    data = get_number(5)

    data.sort()
    print(f"Your Final Results is {data}")

main()
```

```python
#19

numrev = []
for i in range(1,11):
    revin = int(input(f"Please enter your number round {i} : "))
    numrev.append(revin)
print(f"Your current number is {numrev}")

sortnum = sorted(numrev)
print(sortnum)

max_numrev = numrev[0]
for i in numrev:
    if i > max_numrev:
        max_numrev = i
print(f"your current maximum number is {max_numrev}")

acculator = 0
for j in numrev:
    acculator += j
print(f"summation {acculator}")

avg = acculator / len(numrev)
print(f"average {avg}")

moreavg = []
for k in numrev:
    if k > avg:
        moreavg.append(k)
print(f"The number that more than average value is {moreavg}")
```
