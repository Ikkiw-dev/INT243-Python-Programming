```python
#3 logest increasing streak
nums = [1, 2, 3, 2, 3, 4, 5, 1]
# Using a for loop, find: 
# The longest consecutive increasing sequence and Its length
# Expected : Logest streak: [2, 3, 4, 5] Length: 4
max_st = []
cur_st = [nums[0]]

for i in range(1, len(nums)):
    if nums[i] > nums[i-1]:
        cur_st.append(nums[i])
    else:
        if len(cur_st) > len(max_st):
            max_st = cur_st
        cur_st = [nums[i]]
if len(cur_st) > len(max_st):
    max_st = cur_st

print("Logest streak:", max_st)
print("Length:", len(max_st))
```

```python
# Longest Increasing & Decreasing Subsequence
# รับ list จำนวนเต็ม n ตัว
# หาทั้ง:
# Longest Increasing Consecutive Streak
# Longest Decreasing Consecutive Streak
# Input
# 8
# 1 2 3 2 1 0 4 5
# Output
# Increasing: 3
# Decreasing: 4
n = int(input())
nums = list(map(int, input().split()))
max_inc = 1
max_dec = 1
inc_count = 1
dec_count = 1
for i in range(1, n):
    if nums[i] > nums[i-1]:
    #เพิ่ม increasing
        inc_count += 1
        dec_count = 1
    elif nums[i] < nums[i-1]:
    #ลด decreasing
        dec_count += 1
        inc_count = 1
    else:
        inc_count = 1
        dec_count = 1
    if inc_count > max_inc:
        max_inc = inc_count
    if dec_count > max_dec:
        max_dec = dec_count
print("Increasing:", max_inc)
print("Decreasing:", max_dec)
```

```python
# # Character Frequency with Sorting Rules
# รับ string หนึ่งบรรทัด
# ให้แสดงความถี่ตัวอักษรโดยเรียงตาม:
# ความถี่มาก → น้อย
# ถ้าความถี่เท่ากัน → เรียงตามตัวอักษร
text = input()
freq = {}
for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1
items = list(freq.items())
# manual sort (bubble style)
for i in range(len(items)):
    for j in range(i+1, len(items)):
        if items[i][1] < items[j][1]:
            items[i], items[j] = items[j], items[i]
        elif items[i][1] == items[j][1]:
            if items[i][0] > items[j][0]:
                items[i], items[j] = items[j], items[i]
for char, count in items:
    print(char, count)
```

```python
# 💣 ข้อ Bonus (ระดับ Final Exam)
# Mixed Logic Problem
# รับตัวเลข n 
# ให้หาจำนวนตัวเลขตั้งแต่ 1 ถึง n ที่:
# หารด้วย 3 ลงตัว
# มีจำนวนตัวหารเป็นจำนวนคู่
# และผลรวมของ digits เป็นเลข prime
# Input : 50, Output : 5
import math
n = int(input())
result = 0
for i in range(1, n+1):
    # เงื่อนไขที่ 1: หาร 3 ลงตัว
    if i % 3 == 0:
        # เงื่อนไขที่ 2: จำนวนตัวหารเป็นเลขคู่
        # (เท่ากับว่า i ต้องไม่เป็น perfect square)
        if int(math.sqrt(i))**2 != i:
            # เงื่อนไขที่ 3: digit sum เป็น prime
            s = 0
            temp = i
            while temp > 0:
                s += temp % 10
                temp //= 10
            # เช็ค prime
            if s >= 2:
                prime = True
                for j in range(2, int(math.sqrt(s)) + 1):
                    if s % j == 0:
                        prime = False
                        break
                if prime:
                    result += 1
print(result)


```

```python
# 🟢 Problem 5 — Reverse a List (No slicing)
nums = [10, 20, 30, 40, 50]
# Using a while loop, create a new reversed list.
# Do NOT use [::-1].
reversed_list = []
i = len(nums) - 1
while i >= 0:
    reversed_list.append(nums[i])
    i -= 1
print(reversed_list)

```

```python
# 🟡 Problem 6 — Remove Duplicates
nums = [1,1,2,2,3,3,4,4,5]
# Using a while loop, remove duplicate values 
# so that each number appears only once.
unique = []
i = 0
while i < len(nums):
    if nums[i] not in unique:
        unique.append(nums[i])
    i += 1
print(unique)

```

```python
# 🔴 Problem 7 — Two Pointer Technique
nums = [1,2,3,4,5,6,7]
# Using a while loop (no nested loops), 
# determine whether any two numbers sum to 10.
left = 0
right = len(nums) - 1
found = False
while left < right:
    total = nums[left] + nums[right]
    if total == 10:
        found = True
        print("Pair found:", nums[left], nums[right])
        break
    elif total < 10:
        left += 1
    else:
        right -= 1
if not found:
    print("No pair found")

    
```

```python
# 🟢 Problem 8 — Highest Score
scores = {
    "Alice": 88,
    "Bob": 95,
    "Charlie": 78,
    "David": 95
}
# Using a for loop, find:
# The highest score
# All students who have that highest score
max_score = max(scores.values())
top_students = []
for name in scores:
    if scores[name] == max_score:
        top_students.append(name)
print("Highest score:", max_score)
print("Top students:", top_students)

```

```python
🟡 Problem 9 — Group by Grade
scores = {"A": 91,"B": 85,"C": 72,"D": 68,"E": 99}
# Create a new dictionary grouped by grade:
# A → 90+
# B → 80–89
# C → 70–79
# D → below 70
# Expected structure:{"A": [...],"B": [...],"C": [...],"D": [...]}
# Dynamic grading rules (grade : minimum score)
grade_rules = {"A": 90,"B": 80,"C": 70,"D": 0}
grades = {}
# Initialize dictionary dynamically
for grade in grade_rules:
    grades[grade] = []
for student, score in scores.items():
    for grade, min_score in grade_rules.items():
        if score >= min_score:
            grades[grade].append(student)
            break   # Stop once matched
print(grades)

```

```python
# 🔴 Problem 10 — Nested Dictionary
students = {
    "John": {"math": 90, "eng": 85},
    "Anna": {"math": 88, "eng": 92},
    "Mike": {"math": 70, "eng": 75}
}
# Using loops:
# Calculate the average score for each student
# Find the student with the highest average
averages = {}
highest_avg = 0
top_student = ""
for name in students:
    scores = students[name]
    total = 0
    for subject in scores:
        total += scores[subject]
    avg = total / len(scores)
    averages[name] = avg
    if avg > highest_avg:
        highest_avg = avg
        top_student = name
print("Averages:", averages)
print("Top student:", top_student)

```

```python
# 🔴 Problem 12 — Merge Two Dictionaries (Manual)
d1 = {"a":1, "b":2}
d2 = {"b":3, "c":4}
# Merge them so that:
# If a key exists in both dictionaries, add the values
# Expected output:
# {"a":1, "b":5, "c":4}
merged = {}
# Add d1
for key in d1:
    merged[key] = d1[key]
# Merge d2
for key in d2:
    if key in merged:
        merged[key] += d2[key]
    else:
        merged[key] = d2[key]
print(merged)


```
