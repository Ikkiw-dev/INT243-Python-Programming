```python
# ข้อ 2 ใช้ while loop พิมพ์เลข คู่ตั้งแต่ 2 ถึง 20
i = 1
while i <= 20:
    if i %2 == 0:
        print(i)
    i +=1 
```

```python
# ข้อ 3 รับค่าจำนวนเต็ม 1 ค่า แล้วใช้ loop พิมพ์
user = int(input("What is your input?"))
mark = '*'

for i in range(1, user+1):
    print(mark * i)
```

```python
#7 รับ string มา 1 ค่า ตรวจหา a e i o u ใน string
strw = input("What is your preferred words?")
lsw = strw.lower()

a = "a"
e = "e"
i = "i"
o = "o"
u = "u"

countall = 0
counta = 0
counte = 0
counti = 0
counto = 0
countu = 0

for char in strw:
    if char in a:
        counta += 1
        countall += 1
    elif char in e:
        counte += 1
        countall += 1
    elif char in i:
        counti += 1
        countall += 1
    elif char in o:
        counto += 1 
        countall += 1
    elif char in u:
        countu += 1
        countall += 1
    else:
        countall += 0
print(f"Vowels detected : {countall}, a : {counta}, e : {counte}, i :{counti}, o : {counto}, u : {countu}")
```
