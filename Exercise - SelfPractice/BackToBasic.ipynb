```python
#Moodle 1 Review
#1 Univerity grading process
Midterm = int(input("What is your Midterm Score?"))
Final = int(input("What is your Final Exam Score?"))
Project = int(input("What is your project score?"))

if (Midterm > 100 or Midterm < 0) or (Final > 100 or Final < 0) or (Project > 100 or Project < 0):
    print("Invalid Score")
elif (Final < 20):
    print("F")
else:
    score = (0.3*Midterm) + (0.4*Final) + (0.3*Project)
    print(score)

    if score >= 80:
        print("A")
    elif 70 <= score < 80:
        print("B")
    elif 60 <= score < 70:
        print("C")
    elif 50 <= score < 60:
        print("D")
    else:
        print("F")
```

```python
#2 Calculates their sum and difference 
# Prints "Accept" if the sum is greater than 20 AND the difference is even
# Otherwise prints "Reject"

num1 = int(input("What is your first number?"))
num2 = int(input("What is your second number?"))

summation = num1 + num2
differ = num1 - num2
if (summation > 20) and (differ %2 == 0):
    print("Accept")
else:
    print("Reject")
```

```python
#3 Write a script to classify age
# "Child" if age < 13, "Teen" id age < 20, otherwise "Adult"

ages = [12,18,25,15]
for age in ages:
    if age < 13:
        print("Child")
    elif age < 20:
        print("Teen")
    else:
        print("Adult")
```

```python
#4 Write a program to use for loop : Calculates the sum less than 10 and print the sum
numbers = [3,10,15,4,8]
acc = 0
for num in numbers:
    if num < 10:
        acc += num
print(acc)
```

```python
#5 write a program that add 5 to the price if price less that 15 -> otherwise subtracts 2 : print the updates dictionary
product = {'name' : 'pen', 'price' : 17}

if product['price'] < 15:
    product['price'] += 5
else:
    product['price'] -= 2
print(product)
```

```python
#6 Write a python script to calculate the average mark and print the avarage

scores = {'a' : 70, 'b' : 85, 'c': 60}

acc = 0
for i in scores:
    acc += scores[i]
print(acc)

avg = round(acc / len(scores))
print(avg)
```

```python
#8 use a for loop to calculate of numbers less than 10 
lst = [3,10,15,4,8]
total = 0

for i in lst:
    num = int(i)

    if num < 10:
        total += num
print(total)
```

```python
#10 give input score, write a python to seperate the value
# "Distinction" if score >= 80
# "Pass" if score >= 50
# "Fail" in otherwise case

user_in = input("What's your score?")
score = float(user_in)

if score >= 80:
    print("Distinction")
elif score >= 50:
    print("Pass")
else:
    print("Fail")
```
