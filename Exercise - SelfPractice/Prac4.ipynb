```python
#longest Increasing Consecutive Subsequence
nums = [1, 2, 3, 1, 2, 3, 4, 0]

current_length = 1
max_length = 1

for i in range(1, len(nums)):
    if nums[i] > nums[i-1]:
        current_length += 1
    else:
        current_length = 1

    if current_length > max_length:
        max_length = current_length
print(max_length)
```

```python
data = {
    "a": 1,
    "b": 2,
    "c": 1
}

reversed_dict = {}

for key in data:
    value = data[key]

    if value not in reversed_dict:
        reversed_dict[value] = [key]
    else:
        reversed_dict[value].append(key)

print(reversed_dict)
```

```python
# Problem 1 — Zigzag Vertical Print
n = 3
words = ["apple", "banana", "cat"]
#Print vertically, but:
	# •	Even rows → left to right
	# •	Odd rows → right to left
#Expected Output :
# a b c
# t n a
# p a t
# ...

max_word = 0
for w in words:
    if len(w) > max_word:
        max_word = len(w)

for i in range(max_word):
    line = ""

    if i % 2 == 0:
        col_range = range(n)
    else:
        col_range = range(n-1, -1, -1)
        
    for j in col_range:
        if i < len(words[j]):
            line += words[j][i]
        else:
            line += " "
        line += " "
        
    print(line.strip())
```

```python
# Problem 2 - Weighted Alphabet summations
text = "Midterm"
# For each letter:
# 	•	Alphabet value (a=1, z=26)
# 	•	Multiply by (position index + 1)
# 	•	If uppercase → double the value

# Return total sum.
sum_all = 0
alpha = "abcdefghijklmnopqrstuvwxyz"

for i in range(len(text)):
    ch = text[i]
    value = alpha.index(ch.lower()) + 1
    weight = i + 1
    total = value * weight
    if ch != ch.lower():
        total *= 2
    sum_all += total
print(sum_all)
```
