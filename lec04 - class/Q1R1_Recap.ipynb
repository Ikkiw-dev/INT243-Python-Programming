```python
#1 temp compare
temp_a = int(input("What is your Temperature?"))
temp_b = int(input("What is your differ Temperature?"))

temp_b_fah = (temp_b *9/5) + 35

if temp_a > temp_b_fah:
    print("Higher")
elif temp_a < temp_b_fah:
    print("Lower")
else:
    print("Same")
```

```python
#2 Grading controller systems
midterm = int(input("What is your midterm score"))
final_exam = int(input("What is your final score"))
project = int(input("What is your project score"))

if (midterm < 0 or midterm > 100) or (final_exam < 0 or final_exam > 100) or (project < 0 or project > 100):
    print("Error: Invalid Score")
else:
    if final_exam < 20:
        print("F")
    else:
        score = (0.3 * midterm) + (0.4 * final_exam) + (0.3 * project)

        if score >= 80:
            print("A")
        elif 70 <= score < 80:
            print("B")
        elif 60 <= score < 70:
            print("C")
        elif 50 <= score < 60:
            print("D")
        else:
            print("F")
```

```python
#Q1
in1 = input("What is your number input 1?")
in2 = input("What is your number input 2?")

Num1 = int(in1)
Num2 = int(in2)

Summation = Num1 + Num2
differ = Num1 - Num2

if Summation >= 20 and differ %2 == 0:
    print("Accept")
else:
    print("Reject")
```

```python
#Q2
testdict = {'a':70, 'b':85,'c':60}

total = 0
for i in testdict:
    total += testdict[i]

avg = round(total / len(testdict))
print(f"Average : {avg}")
```

```python
#Q9 Write a python script that use a while loop, prints values greater than or equal to 20
values = [5,18,25,7,30]
i = 0
while i < len(values):
    if int(values[i]) > 20:
        print(values[i], end = " ")
    i += 1
```
