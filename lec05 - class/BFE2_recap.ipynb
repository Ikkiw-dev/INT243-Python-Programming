```python
#4 neon sign
n = 3
words = ['hi', 'Ikkyu', 'Pongsatorn']

max_word = 0
for w in words:
    if len(w) > max_word:
        max_word = len(w)

for i in range(max_word):
    line = ""
    for j in range(n):
        if i < len(words[j]):
            line += words[j][i]
        else :
            line += " "

        if j < n - 1:
            line += " "
    print(line)
```

```python
#6 Sum of Alphabet
alpha = "abcdefghijklmnopqrstuvwxyz"
text = "Tutor"

sum_all = 0

for i in range(len(text)):
    ch = text[i].lower()
    if ch in alpha:
        sum_all += alpha.index(ch) + 1 - i
print(sum_all)
```

```python
#7 Factor Calculation
import math
q = int(input("Please key some numbers"))

numbers = list(map(int, input().split()))
results = []

for n in numbers:
    count = 0
    for i in range(1, int(math.sqrt(n)) + 1):
        if n % i == 0:
            count += 1
            if i != n // i:
                count += 1
    results.append(count)
print(results)
```
