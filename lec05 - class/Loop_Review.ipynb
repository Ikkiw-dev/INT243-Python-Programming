```python
#4 factorial

n = int(input("Number : "))

ans = 1
for i in range(1, n+1):
    ans *= i
print(ans)
```

```python
#5 factorial (while)
n = int(input("Number : "))

ans = 1
i = 1
while i <= n:
    ans *= i
    i += 1
print(ans)
```

```python
#13 Triangular Basic

user = int(input("What's your preferred number?"))
for i in range(1, user + 1):
    for j in range(i):
        print("*", end = " ")
    print()
```

```python
#14 Triangular Advance

user = int(input("What's your preferred number?"))
for i in range(user, 0, -1):
    for j in range(i):
        print("*", end = " ")
    print()
```

```python
#15 Numerical Triangular
user_in = int(input("What's your preferred number?"))

for i in range(1, user_in + 1):
    for j in range(1, i + 1):
        print(j, end = " ")
    print()
```

```python
#16 Pyramid printing

user= int(input("What's your preferred number?"))
for i in range(1, user + 1):
    for j in range(user - i):
        print(" ", end = " ")

    for k in range(1, 2 * i):
        print("*", end = " ")
    print()
```
