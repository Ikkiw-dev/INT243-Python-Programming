```python
#1 Reverse neon nvertical sign
n = 2
words = ["hello", "world"]

max_words = 0
for w in words:
    if len(w) > max_words:
        max_words = len(w)

for i in range(max_words - 1, -1, -1):
    line = ""
    for j in range(n):
        if i < len(words[j]):
            line += words[j][i]
        else:
            line += " "

        if j < i + 1:
            line += " "
    print(line)
```

```python
#2 Alphabet difference Score
alpha = "abcdefghijklmnopqrstuvwxyz"
text = "Tutor"

sum_all = 0
for i in range(len(text)):
    ch = text[i].lower()
    if ch in alpha:
        sum_all += alpha.index(ch) + 1 - i
print(sum_all)
```

```python
#3 Count synmetric pairs
text = "level"
count_dup = 0

for i in range(len(text) // 2):
    if text[i] == text[len(text) - 1 - i]:
        count_dup += 1
print(count_dup)
```

```python
#4 Longest Unique Segment
text = "abcabcde"
max_len = 0
for i in range(len(text)):
    seen = ""
    for j in range(i, len(text)):
        if text[j] not in seen:
            seen += text[j]
            max_len = max(max_len, len(seen))
        else:
            break
print(max_len)
print(seen)
```
