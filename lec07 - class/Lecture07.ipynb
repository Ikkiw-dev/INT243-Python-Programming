```python
def analyze_grades(**kwargs):
    total = sum(kwargs.values())
    count = len(kwargs)

    avg = total / count
    highest_score = max(kwargs, key=kwargs.get)
    lowest_score = min(kwargs, key=kwargs.get)

    print(f"Average Score: {avg}")
    print(f"Highest Score: {kwargs.highest_score}")
    print(f"Lowest Score: {kwargs.lowest_score}")
print(analyze_grades(Alice=78, Bob=45, Charlie=90, David=50))
```

```python
#Exercise kwargs 2
def checkout(**kwargs):
    total = 0
    for item, values in kwargs.items():
        amount = values[0]*values[1]
        total += amount

        print("{}:{}".format(item,amount))
    print("Total: {}".format(total))

    #if final total > 1000

    if total >= 1000:
        total *= 0.9

    print(f"Final price is: {total}")
    

checkout(apple=(300,5), milk=(50,3), bread=(40,2))
```

```python
#map function
def power(a,b=2):
    return a**b

result = map(power, [1,2,3,4])
print(result)
print(list(result))

#OR
result = map(lambda a,b=2:a**b, [1,2,3,4])
print(result)
print(list(result))
```

```python
#Exercise map function : convert usd to thb currency with 3 decimal points
def THBconvertion(usd):
    return round(usd*31.5,3)

result = map(THBconvertion, [100,2,50,3])
print(result)
print(list(result))

#
thb_lst = [100,2,50,3]
converted_currency = map(lambda usd: round(usd*31.5,2),thb_lst)
print(list(converted_currency))
```
