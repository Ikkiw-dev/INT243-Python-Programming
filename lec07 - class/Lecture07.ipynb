```python
def compare(num1,num2):
    if num1 > num2:
        print(f"{num1} is larger than {num2}")
    elif num2 > num1:
        print(f"{num2} is larger than {num1}")
    else:
        print("Both number are equal")

compare(1,2)
```

```python
def list_manipulation(lst,cmd,val=None,index=None):
    default_command = ['add','remove','modify']
    cmd = str(cmd)

    if cmd not in default_command:
        return "This Command does not exist."
    elif cmd == "add":
        lst.append(val)
    elif cmd == "remove":
        lst.pop()
    elif cmd == "modify":
        lst[index] = value

    return lst
print(list_manipulation([1,2,3,4],'add',10))
```

```python
def show_items(*args):
    return args
print(show_items(1,2,3,4))
a = [1,2,3,4]
print(show_items(a)) #get : ([1, 2, 3, 4],)
```

```python
def sum_items(*args):
    total = 0
    for i in args:
        total += i
    return total
print(sum_items(1,2,3,4)) #get : 10
```
