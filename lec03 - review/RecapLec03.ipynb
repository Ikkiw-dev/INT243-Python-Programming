```python
password = input("Password please : ")

if len(password) < 8:
    print("ACCESS DENIED")
elif " " in password.lower():
    print("ACCESS DENIED")
elif "secret" not in password.lower():
    print("ACCESS DENIED")
else:
    letter = "abcdefghijklmnopqrstuvwxyz"
    num = "0123456789"

    counttext = 0
    countnum = 0

    for text in password:
        if text.lower() in letter:
            counttext += 1
        elif text in num:
            countnum += int(text)
    
    if counttext > 6 and countnum == 43:
        print("ACCESS GRANTED")
    else:
        print("ACCESS DENIED")
```
